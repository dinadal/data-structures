## **Stack**


* Last In First Out(LIFO)

* **Operations:** push(x),pop()

* **Application:** Reverse a word
image.png

### Example


````
def reverse(s):
 
  # create an empty string
  # stack
  stc = []
 
  # create an empty temporary
  # string
  temp = ""
 
  # traversing the entire string
  for i in range(len(s)):
  
    if s[i] == ' ':
       
      # push the temporary variable
      # into the stack
      stc.append(temp)
       
      # assigning temporary variable
      # as empty
      temp = ""         
 
    else:
      temp = temp + s[i]
 
  # for the last word of the string
  stc.append(temp)
 
  while len(stc) != 0:
 
    # Get the words in reverse order
    temp = stc[len(stc) - 1]
    print(temp, end = " ")
    stc.pop()
   
  print()
 

# This code is contributed by divyeshrabadiya07
s = input("Enter a sentence: ")
reverse(s)
 
````

## Deque


* Double-ended queue
* Stack + Queue
* **Operations:** addFirst(x), removeFirst(), removeLast()
* **Stack:** addFirst(x) + removeFirst()
* **Queue:** addFirst(x) + removeLast()

### Example

````
from collections import deque
     
# Declaring deque
de = deque(['name','age','DOB']) 
de.appendleft(6)
de.append(4)

print(de)
de.popleft()
print(de)
de.pop()
print(de)
````
